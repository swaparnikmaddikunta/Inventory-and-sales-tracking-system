```python
#importing time library --->to print use time.ctime()
import time

#Reading the inventory
fd=open('inv.txt','r')
products=fd.read().split('\n')
fd.close()

#Taking customer details
name=input("Enter customer name=")
phone_no=input("Enter phone number of the customer=")
mail=input("Enter mail id=")

#Taking product details from the customer
user_id=input("Enter the id of the product =")
user_qn=input("Enter the quantity=")

#Taking updated product lst to update inventory which is empty initially
updated_product_lst=[]

for i in products:
    #Getting details of each and every product
    product_details=i.split(',')
    #Checking if the product is available in our inventory
    if(user_id==product_details[0]):
        #If the quantity available in store is more than the need of customer 
        if(int(user_qn)<=int(product_details[3])):
            print('--------------------------------------------------')
            print("Name of the product    =",product_details[1])
            print("Price of the produuct  =",product_details[2])
            print("Qunatity               =",user_qn)
            print('--------------------------------------------------')
            print('Billing amount         =',int(user_qn)*int(product_details[2]))
            print('--------------------------------------------------')

            #Generating sales
            fd=open("Sales.txt","a")
            sales_details=name+','+phone_no+','+mail+','+product_details[0]+','+product_details[1]+','+product_details[2]+','+user_qn+','+str(int(user_qn)*int(product_details[2]))+','+time.ctime()+'\n'
            fd.write(sales_details)
            fd.close()
            
            product_details[3]=str(int(product_details[3])-int(user_qn))
          #If the quantity available in store is less than the need of customer
        else:
            #Asking whether customer is interested in buying the available quantity in the store
            print("Sorry we are having only",product_details[3],"in our store")
            print("Do you want to purchase?")
            print("Press y/Y for yes and any other key for no")
            ch=input("Your key=")
            #If he/she is interested
            if(ch=='y' or ch=='Y'):
                print('--------------------------------------------------')
                print("Name of the product    =",product_details[1])
                print("Price of the produuct  =",product_details[2])
                print("Qunatity               =",product_details[3])
                print('--------------------------------------------------')
                print('Billing amount         =',int(product_details[3])*int(product_details[2]))
                print('--------------------------------------------------')

                fd=open("Sales.txt","a")
                sales_details=name+','+phone_no+','+mail+','+product_details[0]+','+product_details[1]+','+product_details[2]+','+product_details[3]+','+str(int(product_details[3])*int(product_details[2]))+','+time.ctime()+'\n'
                fd.write(sales_details)
                fd.close()
                
                product_details[3]='0'
            #If they are not interested
            else:
                print("Thanks & Visit Again!!")
#Updating the inventory after suitable transactions made         
    updated_product_lst.append(product_details)
print(updated_product_lst)

lst=[]
#Converting updated product list into required format
for i in updated_product_lst:
    if int(i[0])<=3:
        tmp=i[0]+','+i[1]+','+i[2]+','+i[3]+'\n'
    else:
        tmp=i[0]+','+i[1]+','+i[2]+','+i[3]
    lst.append(tmp)
print("list=",lst)
#Making those updated changes in the inventory store
fd=open('inv.txt','w')
for i in lst:
    fd.write(i)
fd.close()
```
