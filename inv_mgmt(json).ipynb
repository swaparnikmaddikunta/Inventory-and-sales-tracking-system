```python
#Original record(Storing data in the form of dictonary)
#record={1001:{'Name':'Five-Star',     'Price':5,    'Qn':200},
#        1002:{'Name':'Bar-One',       'Price':10,   'Qn':150},
#        1003:{'Name':'Candy',         'Price':2,    'Qn':1000},
#        1004:{'Name':'Choclate-cake', 'Price':500,  'Qn':80},
#        1005:{'Name':'Black-Forest',  'Price':550,  'Qn':50}}

#Importing the libraries
import json
import time

#Initiasing sales string as empty
sales=''

#Rading data from json file available in our local system
fd=open('Records.json','r')
js=fd.read()
fd.close()

#Loading the read file into record after converting back into original dictionary format by using "LOADS" inbuilt function
record=json.loads(js)

#Printing the Menu
print('---------------MENU---------------')
for i in record.keys():
    print(i,':',record[i]['Name'],'|',record[i]['Price'],'|',record[i]['Qn'])
print('----------------------------------')

#Taking user-input (about details of customer as well as product)
username=input("Enter the name of the customer=")
mailid=input('Enter the mail id of the user=')
phno=input('Enter phone number=')
user_id=str(input('Enter the id of the product that you want to purchase='))
user_qn=int(input('Enter the quantity of the product that you want to purchase='))

#Calculating Billing Amount if quantity is available in our inventory
if (record[user_id]['Qn']>=user_qn):
    print('----------------------------------')
    print('                BILL              ')
    print('Name of the product  =',record[user_id]['Name'])
    print('Cost of the product  = Rs.',record[user_id]['Price'])
    print('Qunantity            =',user_qn)
    print('----------------------------------')
    print('Total Amount         = Rs.',record[user_id]['Price']*user_qn)
    print('----------------------------------')

    #Generating Sales
    sales_details=username+','+mailid+','+phno+','+user_id+','+record[user_id]['Name']+','+str(record[user_id]['Price'])+','+str(user_qn)+','+str(record[user_id]['Price']*user_qn)+','+time.ctime()+'\n'
    #Updating inventory
    record[user_id]['Qn']=record[user_id]['Qn']-user_qn

#When required quantity for user is not available in the store 
else:
    #Asking whether user wants to purchase the available quantity in store
    print("Sorry,Only",record[user_id]['Qn'],"quantity is available in the store")
    print("Do you still wanted to purchase?\nPress y/Y for yes and any other key for no")
    ch=input("Enter your choice = ")

    #If user wants to purchase
    if (ch=='y' or ch=='Y'):       
        print('----------------------------------')
        print('                BILL              ')
        print('Name of the product  =',record[user_id]['Name'])
        print('Cost of the product  = Rs.',record[user_id]['Price'])
        print('Qunantity            =',record[user_id]['Qn'])
        print('----------------------------------')
        print('Total Amount         = Rs.',record[user_id]['Price']*record[user_id]['Qn'])
        print('----------------------------------')
        sales_details=username+','+mailid+','+phno+','+user_id+','+record[user_id]['Name']+','+str(record[user_id]['Price'])+','+str(record[user_id]['Qn'])+','+str(record[user_id]['Price']*record[user_id]['Qn'])+','+time.ctime()+'\n'
        record[user_id]['Qn']=0
        
    #If user don't want to purchase
    else:
        print("Thank you!! Visit again")
        

#Storing updated inventory in String format in json file
js=json.dumps(record)
fd=open('Records.json','w')
fd.write(js)
fd.close()

#Writing generated sales on to the text file
fd=open('sales.txt','a')
fd.write(sales_details)
fd.close()
```
